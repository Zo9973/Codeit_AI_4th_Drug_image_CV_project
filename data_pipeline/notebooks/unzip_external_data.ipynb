```python
import zipfile
import os
from pathlib import Path
```

```python
def quick_extract_flat(zip_path, output_dir="./data/train_images"):
    """빠른 추출 함수 (확인 과정 없음)"""

    os.makedirs(output_dir, exist_ok=True)
    extracted_count = 0

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        for file_path in zip_ref.namelist():
            # 폴더 건너뛰기
            if file_path.endswith('/'):
                continue

            filename = os.path.basename(file_path)

            # PNG이면서 index.png가 아닌 파일만
            if filename.lower().endswith('.png') and not filename.lower().endswith('index.png'):

                # 파일 데이터 읽기
                file_data = zip_ref.read(file_path)

                # 중복 파일명 처리
                output_path = os.path.join(output_dir, filename)
                counter = 1
                original_filename = filename

                while os.path.exists(output_path):
                    name, ext = os.path.splitext(original_filename)
                    filename = f"{name}_{counter}{ext}"
                    output_path = os.path.join(output_dir, filename)
                    counter += 1

                # 파일 저장
                with open(output_path, 'wb') as f:
                    f.write(file_data)

                extracted_count += 1

                if extracted_count % 100 == 0:  # 진행상황 표시
                    print(f"추출됨: {extracted_count}개")

    print(f"총 {extracted_count}개 파일이 추출되었습니다.")
    print(f"저장 위치: {output_dir}")
```

```python
# 간단한 사용법
zip_path = "./data/train_images/TS_3_조합.zip"
quick_extract_flat(zip_path)
```

```python
# 간단한 사용법
zip_path = "./data/train_images/TS_1_조합.zip"
quick_extract_flat(zip_path)
```

```python
def quick_annotation_extract(zip_path, output_dir="./data/train_annotations"):
    """빠른 어노테이션 추출 함수"""

    print(f"어노테이션 데이터 추출 중...")

    os.makedirs(output_dir, exist_ok=True)

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(output_dir)

    # 결과 확인
    json_count = 0
    for root, dirs, files in os.walk(output_dir):
        json_count += len([f for f in files if f.lower().endswith('.json')])

    print(f"추출 완료!")
    print(f"위치: {output_dir}")
    print(f"JSON 파일 수: {json_count}개")
```

```python
annotation_zip = "./data/train_annotations/TL_1_조합.zip"
output_dir = "./data/train_annotations"
quick_annotation_extract(annotation_zip, output_dir)
```
